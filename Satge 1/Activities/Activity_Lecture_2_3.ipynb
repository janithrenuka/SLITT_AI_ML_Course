```python
import matplotlib.pyplot as plt

import numpy as np
import math
x = np.arange(0, math.pi*2, 0.05)

y = np.sin(x)
plt.plot(x,y)
plt.xlabel("arange")
plt.ylabel("sine")
plt.title("Sine wave")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as img 
import numpy as np
image = img.imread('dog.png') 
fig,axes =  plt.subplots(2,3,figsize = (7,7))



axes[0][0].imshow(image[:,:,0],cmap='Reds') # Red Channel
axes[0][1].imshow(image[:,:,1],cmap='Greens') # Green Channel
axes[0][2].imshow(image[:,:,2],cmap='Blues') # Blue Channel
axes[1][0].imshow(image[:,:,0],cmap='gray') # Red Channel in gray colour
axes[1][1].imshow(image[:,:,1],cmap='gray') # Green Channel in gray colour
axes[1][2].imshow(image[:,:,2],cmap='gray') # Blue Channel in gray colour

plt.tight_layout() 
plt.show() 

```

```python
# Let's try to see images in one of the folders from the Georgia dataset
import matplotlib.pyplot as plt
import matplotlib.image as img # take image class of matplotlib as img
import numpy as np
fig,axes =  plt.subplots(3,5, figsize = (12,5))
sample = "s10" # change this to get a different folder
images = np.empty([15, 480, 640,3], dtype=float)
for r in range(15):
  if (r+1) < 10:
     zero = "0" # to check if we need to have a trailing zero
  else:
     zero = ""
  st = 'gt_db/'+sample+'/'+zero+str(r+1)+'.jpg' # generate file name
  print(st) # display loaded file as debug
  image = img.imread(st) # load image file
  images[r] = image

  y = int(r / 5) # calculate position in plot
  x = r % 5 # calculate position in plot
  axes[y][x].imshow(image) # set the image

plt.tight_layout()   # spaces out the plots 
plt.show() # display all the plots

```
