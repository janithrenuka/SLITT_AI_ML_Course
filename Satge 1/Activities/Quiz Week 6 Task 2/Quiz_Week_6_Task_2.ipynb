1.1 Import Data

```python
import pandas as pd
url = "https://raw.githubusercontent.com/janithrenuka/SLITT_AI_ML_Course/main/Mobile_Price_Classification-220531-204702.csv"
pf = pd.read_csv(url)
pf.head(5)
```

1.2 Define x, y

```python
x = pf.drop(['price_range'],axis=1)
y = pf['price_range']
print(y)
```

2. sklearn model to train data
test size is 0.25 from all data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

3.1 Import Dependenies

```python
from keras.models import Sequential, load_model
from keras.layers import Dense, Dropout
```

3.2 Build Model

```python
model=Sequential()
model.add(Dense(units=8,activation='relu',input_dim=len(x_train.columns)))
model.add(Dense(units=4,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
model.summary()
```

4. Compile the model

```python
model.compile(loss='binary_crossentropy',optimizer='sgd',metrics='accuracy')
```

5. Fit, train the model

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,batch_size=32)
```

6. Save the weight file

```python
model.save('weights.h5')
```
